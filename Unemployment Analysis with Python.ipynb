Unemployment Analysis with Python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load unemployment data (replace filename with actual dataset downloaded)
# Make sure to replace 'YOUR_FILE_PATH_HERE' with the correct path to your file
df = pd.read_csv('Unemployment in India.csv')

# Basic data inspection and cleaning
print(df.head())
print(df.info())

# Handling missing values, if any
df = df.dropna()

# Remove leading space from ' Date' column and convert to datetime with explicit format
df[' Date'] = df[' Date'].str.lstrip()
df[' Date'] = pd.to_datetime(df[' Date'], format='%d-%m-%Y')


# Exploratory Data Analysis
# Plotting unemployment rate over time
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x=' Date', y=' Estimated Unemployment Rate (%)')  # Adjust column names as per dataset
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.show()

# Visualize distribution of unemployment rates
sns.histplot(df[' Estimated Unemployment Rate (%)'], bins=20)
plt.title('Distribution of Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()

# Investigate Covid-19 impact (filter by date range if dataset permits)
covid_period_df = df[(df[' Date'] >= '2020-03-01') & (df[' Date'] <= '2021-12-31')]  # Example range
plt.figure(figsize=(12,6))
sns.lineplot(data=covid_period_df, x=' Date', y=' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate During Covid-19')
plt.xlabel('Date')
plt.ylabel('Rate (%)')
plt.show()

# Findings summary
print("Key patterns and insights:")
print(covid_period_df.describe())
```
